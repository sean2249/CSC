```python
def toDict(lst):
    out = dict()
    if len(lst)%2==0:
        for idx in range(0, len(lst), 2):
            out[lst[idx]] = lst[idx+1]
    return out
```

```python
def speDataframe(_gg):
    # _gg = _gg.groupby(['column field'])
    _ggS = _gg.size()
    _ggDF = pd.DataFrame(_ggS,columns=['count'])
    _ggDF_sort = _ggDF.sort_values('count', ascending=False)
    
    return _ggDF_sort
```

```python
def errorCorrectExtract(pTable, error_ch):
    champ_pair = pTable[pTable['error']==error_ch]
    _gg = champ_pair.groupby(['corr'])
    _ggS = speDataframe(_gg)
    _ggDict = _ggS.to_dict()['count']
    
    return sorted(_ggDict.items(), key=lambda x:x[1], reverse=True)
```

```python
def read_2char(input_filename, groundTruth_filename, ptn):
    '''
    Read the data (sequence, ground truth) extracted from news
    Return nearby dictionary 
    '''
    preError    = defaultdict(lambda: defaultdict(int))
    preCorrect  = defaultdict(lambda: defaultdict(int))
    postError   = defaultdict(lambda: defaultdict(int))
    postCorrect = defaultdict(lambda: defaultdict(int))
    
    errorDict   = defaultdict(lambda: defaultdict(int))
    correctDict = defaultdict(lambda: defaultdict(int))
    
    bigDict = defaultdict(int)
    
    pTable = {'pre':[],'post':[],'corr':[],'error':[],'pre2':[],'post2':[]}
    
    with open(input_filename, 'r', encoding='utf8') as ip, \
    open(groundTruth_filename, 'r', encoding='utf8') as gtp:
        line_idx = 0
        seq_line = ip.readline().strip('\n')
        dataID, seq = seq_line[:seq_line.find(',')], seq_line[seq_line.find(',')+2:]
        

        gt_lst  = gtp.readline().strip('\n').split(', ')
        while seq!='' or gt_lst!=['']:
#         while line_idx <5:
            gt_lst  = gtp.readline().strip('\n').split(', ')
            gtID, gt = gt_lst[0], gt_lst[1:]

            seq_line = ip.readline().strip('\n')
            dataID, seq = seq_line[:seq_line.find(',')], seq_line[seq_line.find(',')+2:]

            # Avoid some exception of sequence that have unwanted \M
            cntdown = 5
            while dataID != gtID and cntdown!=0:
                seq_line = ip.readline().strip('\n')
                dataID, seq = seq_line[:seq_line.find(',')], seq_line[seq_line.find(',')+2:]
                cntdown -= 1
                
            # If the count of trial above, jump out for bad input data 
            if cntdown==0:
                break

            line_idx += 1    

            for idx in range(0, len(gt), 2):
                corr_ch = str(gt[idx+1])
                error_idx = int(gt[idx]) -1
                try:
                    error_ch = seq[error_idx]
                    pre_ch   = str(seq[error_idx-1]) if error_idx-1>=0 else 'x'
                    post_ch  = str(seq[error_idx+1]) if error_idx+1<len(seq) else 'x'
                    pre2_ch  = str(seq[error_idx-2]) if error_idx-2>=0 else 'x'
                    post2_ch = str(seq[error_idx+2]) if error_idx+2<len(seq) else 'x'
                    
                except IndexError:
                    continue
                '''
                1-1 (Error-pre-correct)
                1-2 (Error-post-correct)
                2-1 (Error-pre)
                2-2 (Error-post)
                '''
                
                if ptn.search(corr_ch) or ptn.search(error_ch):
                    continue
                    
                if ptn.search(pre_ch) or ptn.search(post_ch) or ptn.search(pre2_ch) or ptn.search(post2_ch):
                    continue
            
                pTable['pre'].append(pre_ch)
                pTable['post'].append(post_ch)
                pTable['pre2'].append(pre2_ch)
                pTable['post2'].append(post2_ch)
                pTable['corr'].append(corr_ch)
                pTable['error'].append(error_ch)
                

    print('Done.')
    print('= Processing {} sequences'.format(line_idx))
    pTable = pd.DataFrame(pTable)
    return pTable
```

```python
def read(input_filename, groundTruth_filename):
    '''
    Read the data (sequence, ground truth) extracted from news
    Return dictionary of erros-correction (without any filter applied)
    '''
    bigDict = defaultdict(lambda: defaultdict(int))
    with open(input_filename, 'r', encoding='utf8') as ip, \
    open(groundTruth_filename, 'r', encoding='utf8') as gtp:
        line_idx = 0
        seq_line = ip.readline().strip('\n')
        dataID, seq = seq_line[:seq_line.find(',')], seq_line[seq_line.find(',')+2:]

        gt_lst  = gtp.readline().strip('\n').split(', ')
        while seq!='' or gt_lst!=['']:
            gt_lst  = gtp.readline().strip('\n').split(', ')

            seq_line = ip.readline().strip('\n')
            dataID, seq = seq_line[:seq_line.find(',')], seq_line[seq_line.find(',')+2:]

            # Avoid some exception of sequence that have unwanted \M
            cntdown = 5
            while dataID != gt_lst[0] and cntdown!=0:
                seq_line = ip.readline().strip('\n')
                dataID, seq = seq_line[:seq_line.find(',')], seq_line[seq_line.find(',')+2:]
                cntdown -= 1
            # If the count of trial above, jump out for bad input data 
            if cntdown==0:
                break

            line_idx += 1    

            gt = gt_lst[1:]    
            for idx in range(0, len(gt), 2):
                corr_ch = str(gt[idx+1])
                try:
                    error_ch = str(seq[int(gt[idx])-1])
                except IndexError:
                    continue

                bigDict[error_ch][corr_ch] += 1

    print('Done.')
    print('= Processing {} sequences'.format(line_idx))
    print('= Total Error character {}'.format(len(bigDict)))
    
    return bigDict
```

```python
def filterlow(bigDict, threshold):
    new = {}
    idx = 0
    for ch, cands in bigDict.items():
        seg = list(filter(lambda x:x[1]>=threshold, cands.items()))
        
        if len(seg)==0:
            continue

        new[ch] = dict((x,y) for x,y in seg)
    
    print('Done.')
    print('= Total Error character %d' %len(new))
    return new
```

```python
def filterSpecial(bigDict, ptn):
    new = {}
    for ch, cands in bigDict.items():
        if ptn.search(ch):
            continue
            
        seg = [(c_ch, f_ch) for (c_ch, f_ch) in cands.items() if ptn.search(c_ch) is None]
        
        if len(seg)==0:
            continue
        
        new[ch] = dict((c_ch, f_ch) for (c_ch, f_ch) in seg)
   
    print('Done.')
    print('= Total Error character %d' %len(new))
    return new
```

```python
def outputFreqPair(new_dict, output_name):

    dataMeta = defaultdict(dict)

    n_pairs    = 0
    n_allfreqs = 0
    with open(output_name, 'w', encoding='utf8') as fp:
        fp.write('Error\tCorrect\tCount\n')
        for error_ch, corr_cands in new_dict.items():

            n_cands = len(corr_cands.values())
            n_freqs = sum(corr_cands.values())

            ratio = n_freqs/n_cands
            sorted_cands = sorted(corr_cands.items(), key=lambda x:x[1], reverse=True)
            max_pair = sorted_cands[0]
            
            n_pairs += n_cands
            n_allfreqs += n_freqs

            dataMeta[error_ch] = dict(candsType=n_cands, allFreqs=n_freqs, maxPair=max_pair, ratio=ratio)

            for corr, freq in sorted_cands:
                fp.write('%s\t%s\t%d\n' %(error_ch, corr, freq))

    print ('Done.')
    print ('= Total Error pairs %d' %n_pairs)
    print ('= Total events %d' %n_allfreqs)
    print ('= Return Frequency Information')
    print ('= Write file into %s' %output_name)
    
    return dataMeta
```

```python
def outputMeta(dataMeta, bigDict, token=[]):
    sortPair = [(ch, meta['maxPair'][0], meta['maxPair'][1]) \
                for ch, meta in sorted(dataMeta.items(), key=lambda x:x[1]['maxPair'][1], reverse=True)]

    sortCandsType = [(ch, meta['candsType'], list(bigDict[ch].keys())) \
                     for ch,meta in sorted(dataMeta.items(), key=lambda x:x[1]['candsType'], reverse=True)]

    sortRatio = [(ch, meta['ratio'], meta['allFreqs'], meta['candsType'])\
                for ch, meta in sorted(dataMeta.items(), key=lambda x:x[1]['ratio'], reverse=True)]    
    
    print('Done.')
    print('= Return sortPair,sortCandsType,sortRatio')
    
    if len(token)>=1 and type(token)== str:
        f_pair  = 'sortPair_{}.csv'.format(token)
        with open(f_pair, 'w', encoding='utf8') as fp:
            fp.write('Error\tCorrection\tFrequency\n')
            for (error,corr, freqs) in sortPair:
                fp.write('{}\t{}\t{}\n'.format(error,corr,freqs))

        f_cands = 'sortCands_{}.csv'.format(token)
        with open(f_cands, 'w', encoding='utf8') as fp:
            fp.write('Error\tNumberOfTypes\tCorrectionList\n')
            for (ch, n_types, n_cands) in sortCandsType:
                fp.write('{}\t{}\t{}\n'.format(ch, n_types, ','.join(n_cands)))        

        f_ratio = 'sortRatio_{}.csv'.format(token)
        with open(f_ratio, 'w', encoding='utf8') as fp:
            fp.write('Error\tRatio\tTotalFrequency\tNumberOfTypes\n')
            for (ch, ratio, n_freqs, n_types) in sortRatio:
                fp.write('{}\t{:.2f}\t{}\t{}\n'.format(ch, ratio, n_freqs, n_types))        
        print('= Output {},{},{}'.format(f_pair,f_cands,f_ratio))
    
    return (sortPair, sortCandsType, sortRatio)
```
